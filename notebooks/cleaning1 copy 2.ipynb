**Cleaning Movies Metadata**

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('C:/Coding Projects/Movie-Recommender-Engine/data/movies_metadata.csv')
movies['id'] = pd.to_numeric(movies['id'], errors = 'coerce')


links = pd.read_csv('C:/Coding Projects/Movie-Recommender-Engine/data/links_small.csv')
links = links[links['tmdbId'].notnull()]
links['tmdbId'] = links['tmdbId'].astype(int)

movies = movies[movies['id'].isin(links['tmdbId'])]

id_counts = movies['imdb_id'].value_counts()
```

```python
links['tmdbId'][~links['tmdbId'].isin(movies['id'])]
```

```python
movies[movies['id']==1]
links[links['tmdbId'] == 1]
```

```python
duplicate_ids = id_counts[id_counts > 1]
duplicate_ids
```

```python
duplicate_movies = movies[movies['imdb_id'].isin(duplicate_ids.index)]
duplicate_movies.sort_values(by = ['imdb_id', 'id'])
```

```python
movies.drop_duplicates(subset = ['imdb_id', 'id'], inplace = True) #Drop duplicate entries based on imdb_id and id
```

```python
movies['imdb_id'].value_counts() #Duplicates are removed
```

```python
#Drop rows where id or imdb_id is in an invalid format
movies.dropna(subset = ['id'], inplace = True)
movies.dropna(subset = ['imdb_id'], inplace = True)

movies['movie_id'] = movies['id'].astype(int)
movies.drop(columns = ['id'], inplace = True)
```

```python
#Drop rows where vote_count, vote_average, or title is missing
movies.dropna(subset = ['vote_count', 'vote_average', 'title'], inplace = True)
```

```python
#Rename original_title column to title

movies['title'] = movies['title']
movies.drop(columns = ['original_title'], inplace = True)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda element: eval(element))
```

```python
movies.loc[0, 'genres']
movies['genres'] = movies['genres'].apply(lambda element: [d['name'] for d in element] if isinstance(element, list) else [])
```

```python
#Keep overviews
movies.drop(columns = ['belongs_to_collection', 'budget', 'homepage', 'popularity', 'production_companies', 'production_countries', 'release_date', 'revenue', 'runtime', 'spoken_languages', 'status', 'poster_path', 'video', 'original_language', 'adult'], inplace = True)
```

```python
movies['overview'] = movies['tagline'].fillna('').str.cat(movies['overview'].fillna(''), sep=' ')
movies.drop(columns = ['tagline'], inplace = True)
movies.head()
```

```python
"""Stem the keywords so there are less features in the TF-IDF matrix"""
import string
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer('english')

movies['overview'] = movies['overview'].str.split()
movies['overview'] = movies['overview'].apply(lambda l: [stemmer.stem(word.strip(string.punctuation)) for word in l])
movies.head()
```

```python
#Calculate the imdb score

min_votes = movies['vote_count'].quantile(0.90)
overall_vote_average = movies['vote_average'].mean()

imdb_score = (movies['vote_count'] * movies['vote_average'] + min_votes * overall_vote_average) / (movies['vote_count'] + min_votes)
movies['imdb_score'] = imdb_score

unqualified_movies = movies['vote_count'] < min_votes
movies.loc[unqualified_movies, 'imdb_score'] = np.nan

movies.drop(columns = ['vote_count', 'vote_average'], inplace = True)
```

```python
movies.head()
```

**Cleaning Keywords**

```python
keywords = pd.read_csv('C:/Coding Projects/Movie-Recommender-Engine/data/keywords.csv')
keywords['keywords'] = keywords['keywords'].apply(lambda element: eval(element))
keywords['movie_id'] = keywords['id']
keywords.drop(columns = ['id'], inplace = True)
keywords.head()
```

```python
keywords['keywords'] = keywords['keywords'].apply(lambda element: [d['name'] for d in element] if isinstance(element, list) else [])
keywords.head()
```

```python
all_keywords = keywords['keywords'].apply(lambda l: pd.Series(l)).stack().value_counts()
all_keywords.head()

unique_keywords = all_keywords[all_keywords == 1]
unique_keywords
```

```python
keywords['keywords'] = keywords['keywords'].apply(lambda l: [element for element in l if element not in unique_keywords.index])
keywords.head()
```

```python
"""Stem the keywords so there are less features in the TF-IDF matrix"""
keywords['keywords'] = keywords['keywords'].apply(lambda l: [stemmer.stem(word) for word in l])
```

```python
keywords.head()
```

**Cleaning credits**

```python
credits = pd.read_csv('C:/Coding Projects/Movie-Recommender-Engine/data/credits.csv')
credits.head()
```

```python
credits['cast'] = credits['cast'].apply(lambda element: eval(element))
credits['crew'] = credits['crew'].apply(lambda element: eval(element))
credits.head()
```

```python
credits['director'] = credits['crew'].apply(lambda element: [d['name'] for d in element if 'director' in d['job'].lower()] if isinstance(element, list) else [])

"""Turn the director names into a single string not seperated by spaces (otherwise first and last name will be treated as different terms in TF-IDF vectorization)"""

credits['director'] = credits['director'].apply(lambda l: [''.join(fullname.split()) for fullname in l])
credits['director'].head()
```

```python
"""Find the cast members that only appear in one movie"""

director_name_counts = credits['director'].apply(lambda l: pd.Series(l)).stack().value_counts()
unique_directors = director_name_counts[director_name_counts == 1]
unique_directors.head(30)
unique_directors.info()
```

```python
credits['director'] = credits['director'].apply(lambda l: [name for name in l if name not in unique_directors.index][0:3]) #Take the top 3 directors
credits['director'].head()
```

```python
credits.head()
```

```python
actors = credits['cast'].apply(lambda l: [d['name'] for d in l if 'character' in d.keys() and d['character'] != ''][0:10]) #take the top 10 actors
actors = actors.apply(lambda l: [''.join(fullname.split()) for fullname in l])
actors.head()
```

```python
credits['actors'] = actors
credits.head()
```

```python
credits['movie_id'] = credits['id']
credits.drop(columns = ['id'], inplace = True)
credits.head()
```

```python
#Director should be weighted twice as much as the actors
credits['director'] = credits['director'] * 2
credits.head()
```

**Merge Keywords and Credits into Movies**

```python
movies = movies.merge(keywords, on = 'movie_id', how = 'inner')
movies.head()
```

```python
movies = movies.merge(credits.loc[:, ['movie_id', 'actors', 'director']], on='movie_id', how='inner')
movies['actors'] = movies['actors'].apply(lambda x: x if isinstance(x, list) else [])
movies.head()

movies.info()
```

```python
movies.head()
movies.info()
```

**Create the Cosine Similarity Matrix between genres**

```python
movies.head()
```

```python
genres = movies['genres'].apply(lambda l: ' '.join(l) if isinstance(l, list) else '')
movies.drop(columns = ['genres'], inplace = True)
genres.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(stop_words = 'english', ngram_range = (1, 2), min_df = 0.01)
dtm = tfidf_vectorizer.fit_transform(genres)

genres_sim_matrix = pd.DataFrame(cosine_similarity(dtm, dense_output = True))

genres_sim_matrix.index = movies['movie_id'].values
genres_sim_matrix.columns = movies['movie_id'].values

genres_sim_matrix.head()
```

```python
genres_sims_matrix = genres_sim_matrix.stack().reset_index() #if we were to have used movies['movie_id'] instead of movies['movies_id'].values, the stack and reset_index would create a dataframe where 2 columns/series have the .name as 'movie_id', creating an error
genres_sims_matrix.columns = ['movie1_id', 'movie2_id', 'similarity_score']

genres_sims_matrix.head()
```

**Create the cosine similarity matrix for movie crew**

```python
crew = movies.apply(lambda row: row['director'] + row['actors'], axis=1)
crew = crew.apply(lambda x: ' '.join(x))
                                        
movies = movies.drop(columns = ['director', 'actors'])

crew.head()
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words = 'english', min_df = 2)
dtm = tfidf_vectorizer.fit_transform(crew)

crew_sim_matrix = pd.DataFrame(cosine_similarity(dtm, dense_output = True))
crew_sim_matrix.index = movies['movie_id'].values
crew_sim_matrix.columns = movies['movie_id'].values
```

```python
print(f"Vocabulary size: {len(tfidf_vectorizer.vocabulary_)}")
```

```python
print(movies['movie_id'].name)

crew_sims = crew_sim_matrix.stack().reset_index()
crew_sims.columns = ['movie1_id', 'movie2_id', 'similarity_score']
```

```python
crew_sims[crew_sims['similarity_score'] != 0]
```

```python
#Drop rows where similarity score is 0 (no similarity) or 1 (self similarity)for storage efficiency
crew_sims = crew_sims[(crew_sims['movie1_id'] != crew_sims['movie2_id']) & (crew_sims['similarity_score'] > 0.0)]
crew_sims.head()
```

**Calculate Similarities between movie content**

```python
content = movies.apply(lambda row: row['keywords'] + row['overview'] * 2, axis=1)
content = content.apply(lambda x: ' '.join(x))
                                        
movies = movies.drop(columns = ['keywords', 'overview'])

content.head()

```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words = 'english', min_df = 0.01)
dtm = tfidf_vectorizer.fit_transform(content)

content_sim_matrix = pd.DataFrame(cosine_similarity(dtm, dense_output = True))
content_sim_matrix.index = movies['movie_id'].values
content_sim_matrix.columns = movies['movie_id'].values

content_sim_matrix.head()
```

```python
content_sims = content_sim_matrix.stack().reset_index()
content_sims.columns = ['movie1_id', 'movie2_id', 'similarity_score']

content_sims.head()
```

```python
#Drop rows where similarity score is 0 (no similarity) or 1 (self similarity)for storage efficiency
content_sims = content_sims[(content_sims['movie1_id'] != content_sims['movie2_id']) & (content_sims['similarity_score'] > 0.0)]
content_sims.head()
```

**Create the correlation matrix for movie ratings**

```python
movie_ratings = pd.read_csv(f'C:/Coding Projects/Movie-Recommender-Engine/data/ratings_small.csv')
movie_ratings.head()
```

```python
movie_ratings = movie_ratings.pivot_table(index = 'userId', columns = 'movieId', values = 'rating')
movie_ratings.head()
```

```python
movie_ratings.info()
```

```python
correlation_matrix = movie_ratings.corr(method = 'pearson', min_periods = 7)
correlation_matrix.index.name = None
correlation_matrix.head()
```

```python
correlation_matrix.index.name = 'movie1_id'
correlation_matrix.columns.name = 'movie2_id'
```

```python
ratings_sims = correlation_matrix.stack().reset_index()
ratings_sims.columns = ['movie1_id', 'movie2_id', 'similarity_score']

ratings_sims.head()
```

```python
#Drop rows where similarity score is 0 (no similarity) or 1 (self similarity)for storage efficiency
ratings_sims = ratings_sims[(ratings_sims['movie1_id'] != ratings_sims['movie2_id']) & (ratings_sims['similarity_score'] > 0.0)]

ratings_sims.head()
```
